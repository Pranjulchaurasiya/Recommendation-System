##Popularity Based Recommender system

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Convert 'Book-Rating' to numeric, invalid parsing will be set as NaN
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')

# Group by 'Book-Title' and calculate the mean rating
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the column
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)

# Display the result
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Image-URL-M','Book-Author','num_ratings','avg_rating']]
popular_df
```

```python
print(books.columns)
print(popular_df.columns)
```

```python
popular_df = popular_df.merge(
    books[['Book-Title', 'Book-Author', 'Image-URL-M']],
    on='Book-Title',
    how='left'
)

print(popular_df.head())
print(popular_df.columns)
```

```python
popular_df = popular_df.drop_duplicates('Book-Title')[['Book-Title', 'num_ratings','Book-Author', 'Image-URL-M','avg_rating']]
print(popular_df.head())
```

##Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
import sklearn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
import numpy as np
from rapidfuzz import process, fuzz

def recommend(book_name):
    # Fuzzy match the book name to find the closest match in the index
    closest_match = process.extractOne(book_name, pt.index, scorer=fuzz.token_sort_ratio)
    
    if closest_match is None or closest_match[1] < 70:  # Threshold for similarity
        return f"No close match found for '{book_name}'"
    
    # Extract matched keyword and similarity score
    matched_book, similarity_score = closest_match[0], closest_match[1]
    print(f"🔍 Closest Match Found: '{matched_book}' (Similarity: {similarity_score}%)")
    
    # Fetch index
    index = np.where(pt.index == matched_book)[0][0]
    
    # Fetch top similar items
    similar_items = sorted(
        list(enumerate(similarity_scores[index])),
        key=lambda x: x[1],
        reverse=True
    )[1:6]  # Top 5 similar books, excluding the queried one
    
    # Fetch book details
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
